```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
X=np.linspace(0,10,100)
plt.plot(X)
```

```python
y=np.sin(X)
plt.plot(y)
```

```python
fix,ax=plt.subplots()
ax.scatter(X,y)
ax.set_xlabel("X label")
ax.set_ylabel("Y label")
ax.grid()
plt.show()
```

Additional Learning

```python
X=np.random.randn(100)
y=np.random.randn(100)

fig,ax=plt.subplots()
ax.scatter(X,y)
ax.set_xlabel("X_Axis")
ax.set_ylabel("Y_Axis")
ax.set_title("Scatter Plot")
ax.grid()
plt.show()
```

Bar Graph

```python
X=['a','b','c','d','e','f']
y=[1,2,5,7,8,9]
fig,ax=plt.subplots()
ax.bar(X,y)
plt.show()
```

```python
X=data2['Pclass']
Y=data2['Fare']
fig,ax=plt.subplots()
ax.bar(X,Y)
plt.show()
```

Histogram

```python
X=data2['Age']
fig,ax=plt.subplots()
ax.hist(X)
plt.show()
```

Piechart

```python
X=[15,30,45,10]
Y=['A','B','C','D']
fig,ax=plt.subplots()
ax.pie(X,labels=Y,autopct="%0.2f%%")
plt.show()
```
