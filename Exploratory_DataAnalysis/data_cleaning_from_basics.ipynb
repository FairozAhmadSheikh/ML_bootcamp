```python
import numpy as np
```

```python
datacopy['Size']=datacopy['Size'].str.replace('M','000')
datacopy['Size']=datacopy['Size'].str.replace('k','')
datacopy['Size']=datacopy['Size'].str.replace('+','')
datacopy['Size']=datacopy['Size'].str.replace(',','')
datacopy['Size']=datacopy['Size'].str.replace('Varies with device',str(np.nan))
datacopy['Size']=datacopy['Size'].astype('float')
```
