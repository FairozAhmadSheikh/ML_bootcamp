```python
import pandas as pd
import numpy as np
```

```python
df=pd.DataFrame({
    'A':[1,2,3,4,5,6,100],
    'B':[10,20,30,40,50,60,700],
    'C':[100,200,300,400,500,600,7000]})
df
```

```python
# way 1 
def replace_outliers_with_median(x):
    median=x.median()
    std=x.std()
    lower=median-3*std
    upper=median-3*std
    x[(x<lower)|(x>upper)]=median
    return x
df=df.apply(replace_outliers_with_median)
df
```
