Random Numbers Generation, array generation etc

```python
import numpy as np
```

rand= Generates number between 0 and 1 

```python
val=np.random.rand(5)   #1d
val
```

```python
two_d=np.random.rand(5,4)
two_d
```

randn= Generates number that are normally distributed

```python
arr1=np.random.randn(500)
arr1
```

```python
import seaborn as sns
sns.displot(arr1,kde=True)
```

```python
arr2=np.random.randn(40,50)
sns.displot(arr2,kde=True)
```

Random Integers Generation
randint(start,end,number of elemsnts you want)

```python
values=np.random.randint(20,50,size=20)
print(values)
values.size
```

arange function  generates sequence but stop  point is exclusive 

```python
np.arange(20,30,dtype=float)
```

Choice used for selection random selection 

```python
array1=np.array([1,2,3,4,5,6])
np.random.choice(array1,size=3)   # selects 3 random samples
```

```python
population=np.arange(1,100)
sample=np.random.choice(population,size=10)
sample
```
