```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
population=np.random.binomial(10,0.5,10000)
sns.displot(population)
```

```python
log_normal=np.random.lognormal(10,0.5,10000)
sns.displot(log_normal)
```

Take Sample Now

```python
SIZE=int(len(population)*0.30)
SIZE
```

Samples Now

```python
sample_means=[]
for i in range(1,31):
    sample=np.random.choice(population,size=SIZE)
    sample_means.append(np.mean(sample))
```

Sample Mean Distribution Plot 

```python
sns.displot(sample_means)
```
