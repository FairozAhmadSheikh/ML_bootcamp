qq or quantile quantile plot gives a read line if our data falls on the red line then the data is normally distributed 

```python
d=np.random.lognormal(3,1,1000) 
```

```python
import matplotlib.pyplot as plt
import scipy.stats as stat
import pylab
import seaborn as sns
```

```python
def plot_data(sample):
    plt.Figure(figsize=(10,8))
    plt.subplot(1,2,1)
    sns.histplot(sample,kde=True)
    plt.subplot(1,2,2)
    stat.probplot(sample,dist='norm',plot=pylab)
    plt.show()
plot_data(np.log(d))
```
